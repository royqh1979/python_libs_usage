### basic dummy coding

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np

gfr = pd.read_csv("data/GlastonburyFestivalRegression.dat",sep="\t")

gfr["change"]=gfr["change"].replace('^\s*$',np.nan,regex=True)
gfr["change"]=gfr["change"].astype("float64")
gfr=gfr.dropna()
gfr
```

```python
gfr.describe()
```

##### Use dummy coding
There are 4 groups, and Group "No Musical Affiliation" is the baseline.

```python
import patsy
```

```python
model=smf.ols(formula="change ~ C(music, Treatment('No Musical Affiliation'))",data=gfr)
fit = model.fit()
fit.summary()
```

##### Set dummy coding manually

```python
gfr["music"]=gfr["music"].astype("category")

gfr["music"]
```

Note there are 4 groups ('Crusty','Indie Kid''Metaller''No Musical Affiliation').<br>
And Pay attention to their orders.

```python
code = [
    [1,0,0],
    [0,1,0],
    [0,0,1],
    [0,0,0]
]
```

```python
model=smf.ols(formula="change ~ C(music, code)",data=gfr)
fit = model.fit()
fit.summary()
```
