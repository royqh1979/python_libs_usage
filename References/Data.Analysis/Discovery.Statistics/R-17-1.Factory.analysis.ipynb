## Test 

### Barlett's test

**Bartlett's test** examines where the poplulation correlation matrix resembles an identity matrix. (Varialbes are independant to each other.)

If it's significant, then it means the correlations between variables are significantly different from zero.

```r
library(psych)

# run it from the raw data, we will have an warning.
cortest.bartlett(raqData)
```

```r
# run it from the correlation matrix

nr = nrow(raqData)
cortest.bartlett(raqMatrix, n=nr)
```

p-value < 0, it's significant, so we can do factor analysis.

### KMO index

**KMO (Kaiser-Meyer-Olkin) measure of sampling adequacy** represents the ratio of the squared correlation between variables to the square partial correlation between variables. 
The KMO statistics varies between 0 and 1.
* A value of 0 indicates factor analysis is likely to be inappropriate.
* A value close to 1 indicates that patterns of correlations are relatively copact and so factor analysis should yield distince and reliable factors.
* KMO values greater than 0.5 are barely acceptable
* KMO values between 0.5 and 0.7 are mediocre
* KMO values between 0.7 and 0.8 are good
* KMO values between 0.8 and .0.8 are great
* KMO values above 0.9 are superb (for factor analysis)

```r
library(psych)
KMO(raqData)
```

All values are well above 0.5, so it's ok.

### determinant of the correlation matrix

Finnaly, The determinant of the correlation matrix should be greater than 0.00001.

```r
det(raqMatrix)
```

## Principal components analysis

### First Model

```r
pc1 <- principal(raqData, nfactors = 23, rotate = "none")

summary(pc1)
pc1
```

The **eigenvalues** associated with each factor are called **SS loadings**(Sum of squared loadings) in the result.

Dipslay loadings of variables to the factor. (**Loadings > 0.3 is highlighted**).

```r
library(IRdisplay)

library(htmlTable)
library(magrittr)
# A simple output
cf <-htmlTable(pc1$loadings,css.cell=apply(pc1$loadings,c(1,2),function(v)ifelse( abs(v)>0.3,"color:red","")))

display_html(cf)
```

### Screeplot

```r
plot(pc1$values, type="b")
```

The evidence from the scree plot suggests a four-compontent model.

### Second Model

```r
pc2 <- principal(raqData, nfactors=4, rotate = "none")

pc2
```

```r
library(IRdisplay)

library(htmlTable)
library(magrittr)
# A simple output
cf <-htmlTable(pc2$loadings,css.cell=apply(pc2$loadings,c(1,2),function(v)ifelse( abs(v)>0.3,"color:red","")))

display_html(cf)
```

### Residual Analysis

```r
residuals<-factor.residuals(raqMatrix,pc2$loadings)

residuals <- residuals[upper.tri(residuals)]
```

```r
sum(abs(residuals)>0.5) 

sum(abs(residuals)>0.5) / length(residuals)
```

```r
sqrt(mean(residuals^2))
```

```r
hist(residuals)
```

## Rotation

### Orthogonal Rotation (varimax)

```r
pc3 <- principal(raqData, nfactors = 4, rotate = "varimax")

pc3
```

```r
print.psych(pc4,cut = 0.3, sort=F)
```

### Oblique rotation

```r
pc4<-principal(raqData, nfactors=4, rotate="oblimin")

print.psych(pc4, cut=0.3, sort=T)
```

```r
structure = round(pc4$loading %*% pc4$Phi,2)
library(IRdisplay)

library(htmlTable)
library(magrittr)
# A simple output
cf <-htmlTable(structure,css.cell=apply(structure,c(1,2),function(v)ifelse( abs(v)<0.2,"visibility: hidden","")))

display_html(cf)
```
