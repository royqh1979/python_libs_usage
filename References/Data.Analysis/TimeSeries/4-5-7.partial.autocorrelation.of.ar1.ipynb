产生一个ar(1)序列，绘制其自相关图和局部自相关图 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(0,1,1000)

for i in range(1,1000):
    x[i]=0.7*x[i-1]+x[i]
```

```python
plt.plot(x,"-")
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(x)
```

```python
plot_pacf(x)
```

使用ar()函数估计序列的参数

```python
from statsmodels.tsa.api import AutoReg

model = AutoReg(x,1)
fit=model.fit()
fit.summary()
```
