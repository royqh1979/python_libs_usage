模拟一个harmonic周期序列

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=120
t=np.arange(1,n+1)
w=np.random.normal(0,0.5,n)
trend = 0.1+0.005*t+0.001*t**2
seasonal = np.sin(2*np.pi*t/12)+0.2*np.sin(2*np.pi*2*t/12)+0.1*np.sin(2*np.pi*4*t/12)+0.1*np.cos(2*np.pi*4*t/12)
x = trend+seasonal+w
```

```python
plt.plot(x)
```
