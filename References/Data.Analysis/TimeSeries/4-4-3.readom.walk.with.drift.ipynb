读入HP公司股价序列

```python
import pandas as pd

hp = pd.read_csv("data/HP.txt")
```

```python
hp.head()
```

```python
import numpy as np
```

```python
dp = pd.Series(np.diff(hp["Price"]))
```

```python
dp.head()
```

查看序列的时序图

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dp)
```

绘制序列的自相关图

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(dp)
```

自相关不显著，可以用白噪声模型来描述该序列。

计算白噪声漂移量如下：

```python
np.mean(dp) + np.array([-2,2]) * np.std(dp) / np.sqrt(len(dp))
```
