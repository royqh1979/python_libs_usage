模拟一个Random Walk

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.random.normal(0,1,1000)
```

```python
for t in range(1,1000):
    x[t]=x[t-1]+x[t]
```

```python
plt.plot(x)
```

绘制自相关图

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(x)
```

计算一阶差分序列，并绘制自相关图

```python
plot_acf(np.diff(x))
```
